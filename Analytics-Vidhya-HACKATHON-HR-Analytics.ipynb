```python
train_data['gender'] = Label.fit_transform(train_data['gender'])
train_data['recruitment_channel'] = Label.fit_transform(train_data['recruitment_channel'])
train_data['department'] = Label.fit_transform(train_data['department'])
train_data['region'] = Label.fit_transform(train_data['region'])
train_data['education'] = Label.fit_transform(train_data['education'])
```

```python
X = train_data.iloc[:,1:13]
y = train_data.iloc[:,13]
```

```python
y.head()
```

```python
#from sklearn.ensemble import RandomForestClassifier 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=40)
```

```python
#clf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)
#clf.fit(X_train, y_train)
```

```python
import xgboost as xgb
xgb_model = xgb.XGBClassifier(n_estimators=500, max_depth=5, objective="binary:logistic", random_state=40)
xgb_model.fit(X_train, y_train)
```

```python
y_predicted = xgb_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_predicted)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test, y_predicted, average='macro')
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predicted)
```
